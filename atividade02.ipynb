```python
from openai import OpenAI

client = OpenAI(
  api_key="<key>")

print('Bem-vindo ao ChatBot da Open AI! (Digite x se você quiser sair!)\n')
mensagens = []

while True:
  pergunta = input('O que você precisa?: ')
  if pergunta.lower() == 'x':
    break

  mensagens.append({'role': 'user', 'content': pergunta})

  response = client.responses.create(
    model="gpt-5-nano",
    instructions=(
        "Você é um assistente especializado em gestão de projetos. "
        "Responda sempre em português do Brasil, de forma objetiva. "
    ),
    input=pergunta,
    text={
        "format": {
            "type": "text"
        },
        "verbosity": "low"
    },
    reasoning={
        "effort": "minimal",
        "summary": "auto"
    },
    tools=[],
    store=False
  )

  resposta = response.output_text
  mensagens.append({'role': 'assistant', 'content': resposta})
  print(f"Bot: {resposta}\n")

print('\nMuito obrigado por utilizar o ChatBot da Open AI! O histórico completo da conversa foi:')
print(mensagens)
```
